```python
# AIM = # create 4 cols HDD,SSD,Flask_storage and hybrid 

# Convert 'Memory' column to string
df['Memory'] = df['Memory'].astype(str).replace(r'\.0', '', regex=True)
# remove '.0' and clean GB/TB
df["Memory"] = df["Memory"].str.replace('GB', '').str.replace('TB', '000')

# Split the memory column into two parts if there are multiple storage types
new = df["Memory"].str.split("+", n=1, expand=True)

# Strip whitespace and assign the values to 'first' and 'second'
df["first"] = new[0].str.strip()
df["second"] = new[1].fillna("0").str.strip()  # Ensure 'second' column has no NaNs

# Create new columns to identify storage types in 'first'
df["Layer1HDD"] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

# Clean 'first' column to keep only numeric values
df['first'] = df['first'].str.extract(r'(\d+)')  # Extract numeric part

# Apply the same process for 'second'
df["Layer2HDD"] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

# Clean 'second' column to keep only numeric values
df['second'] = df['second'].str.extract(r'(\d+)')  # Extract numeric part

# Convert 'first' and 'second' to integers for computation
df["first"] = df["first"].fillna(0).astype(int)  # Handle NaNs, convert to int
df["second"] = df["second"].fillna(0).astype(int)

# Calculate the final values for each storage type
df["HDD"] = (df["first"] * df["Layer1HDD"] + df["second"] * df["Layer2HDD"])
df["SSD"] = (df["first"] * df["Layer1SSD"] + df["second"] * df["Layer2SSD"])
df["Hybrid"] = (df["first"] * df["Layer1Hybrid"] + df["second"] * df["Layer2Hybrid"])
df["Flash_Storage"] = (df["first"] * df["Layer1Flash_Storage"] + df["second"] * df["Layer2Flash_Storage"])

# Drop unnecessary columns
df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage'], inplace=True)

# Check the final DataFrame
print(df.head())
```

### Linear regression

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error
import warnings
warnings.filterwarnings('ignore', category=UserWarning, message="Found unknown categories")

# Function to remove missing columns from a list
def get_valid_columns(df, columns):
    return [col for col in columns if col in df.columns]

# Get valid columns for both categorical and numerical columns
valid_categorical_columns = get_valid_columns(X_train, ['Company', 'TypeName', 'ScreenResolution', 'Cpu brand', 'Gpu brand', 'os'])
valid_numerical_columns = get_valid_columns(X_train, ['Ram', 'Weight', 'Touchscreen', 'Ips', 'ppi', 'HDD', 'SSD'])

# Create the ColumnTransformer with handle_unknown='ignore'
step1 = ColumnTransformer(transformers=[
    ('cat', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), valid_categorical_columns)
], remainder='passthrough')  # Remainder 'passthrough' allows numerical columns to pass unchanged

# Define the model
step2 = LinearRegression()

# Create the pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Fit the pipeline to the training data
pipe.fit(X_train, y_train)

# Predict the values for X_test
y_pred = pipe.predict(X_test)

# Evaluate the model
print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
```

### Voting Regressor

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor, VotingRegressor
from xgboost import XGBRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error

# Utility function to get valid columns that exist in the DataFrame
def get_valid_columns(df, columns):
    return [col for col in columns if col in df.columns]

# Check for valid columns in X_train
valid_categorical_columns = get_valid_columns(X_train, ['Company', 'TypeName', 'ScreenResolution', 'Cpu brand', 'Gpu brand', 'os'])
valid_numerical_columns = get_valid_columns(X_train, ['Ram', 'Weight', 'Touchscreen', 'Ips', 'ppi', 'HDD', 'SSD'])

# Optionally print missing columns for debugging purposes
missing_columns = [col for col in ['Company', 'TypeName', 'ScreenResolution', 'Cpu brand', 'Gpu brand', 'os'] if col not in X_train.columns]
if missing_columns:
    print(f"Warning: The following columns are missing and will be ignored: {missing_columns}")

# Create the ColumnTransformer with only valid columns
step1 = ColumnTransformer(transformers=[
    ('cat', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), valid_categorical_columns)
], remainder='passthrough')  # Pass numerical columns through unchanged

# Define individual models
rf = RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5, max_features=0.75, max_depth=15, bootstrap=True)
gbdt = GradientBoostingRegressor(n_estimators=100, max_features=0.5)
xgb = XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5)
et = ExtraTreesRegressor(n_estimators=100, random_state=3, max_samples=0.5, max_features=0.75, max_depth=10, bootstrap=True)

# Voting Regressor with defined weights
step2 = VotingRegressor([('rf', rf), ('gbdt', gbdt), ('xgb', xgb), ('et', et)], weights=[5, 1, 1, 1])

# Create the pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Fit the pipeline to the training data
pipe.fit(X_train, y_train)

# Predict the values for X_test
y_pred = pipe.predict(X_test)

# Evaluate the model
print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
```

### Stacking

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import Ridge
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error

# Utility function to check if columns exist in the dataset
def get_valid_columns(df, columns):
    return [col for col in columns if col in df.columns]

# Check for valid columns in the dataset
valid_categorical_columns = get_valid_columns(X_train, ['Company', 'TypeName', 'ScreenResolution', 'Cpu brand', 'Gpu brand', 'os'])
valid_numerical_columns = get_valid_columns(X_train, ['Ram', 'Weight', 'Touchscreen', 'Ips', 'ppi', 'HDD', 'SSD'])

# Optionally print missing columns for debugging purposes
missing_columns = [col for col in ['Company', 'TypeName', 'ScreenResolution', 'Cpu brand', 'Gpu brand', 'os'] if col not in X_train.columns]
if missing_columns:
    print(f"Warning: The following columns are missing and will be ignored: {missing_columns}")

# Create the ColumnTransformer with only valid columns
step1 = ColumnTransformer(transformers=[
    ('cat', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), valid_categorical_columns)
], remainder='passthrough')  # Pass numerical columns through unchanged

# Define the base estimators for stacking
estimators = [
    ('rf', RandomForestRegressor(n_estimators=350, random_state=3, max_samples=0.5, max_features=0.75, max_depth=15, bootstrap=True)),
    ('gbdt', GradientBoostingRegressor(n_estimators=100, max_features=0.5)),
    ('xgb', XGBRegressor(n_estimators=25, learning_rate=0.3, max_depth=5))
]

# Define the final estimator for StackingRegressor
step2 = StackingRegressor(estimators=estimators, final_estimator=Ridge(alpha=100))

# Create the pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Fit the pipeline to the training data
pipe.fit(X_train, y_train)

# Predict the values for X_test
y_pred = pipe.predict(X_test)

# Evaluate the model
print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
```
